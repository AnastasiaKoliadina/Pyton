#### 4.3 Работа с датасетом "house price"
4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
    - Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
plt.figure(figsize=(8, 6))
corr_matrix = np.round(df.drop(columns=['id','date', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15','yr_renovated']).corr(method='spearman'), 2)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
sns.heatmap(corr_matrix, annot=True, linewidths=1.5,cmap='coolwarm')
```

По матрице корреляции видно большую зависимость цены от жилой площади и меньшую от количества ванных комнат. Однако стоит заметить также корреляцию цены и площади чердака, это говорит от том, что большой популярностью пользуются дома с 1 этажом.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше жилая площадь, тем больше ее стоимость. Но квартир с бо стоимостью и площадью меньше, чем  квартир со средней ценой и площадью

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Конструкция и дизайн')
plt.title('Распределение стоимости в зависимости от дизайна и конструкции');
```

Чем выше оценка дизайна и конструкции тем выше стоимость дома

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Этажи')
plt.title('Распределение стоимости от этажности');
```

Наибольшую популярность и стоимость имеют дома с количеством этажей 2,5

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Стоимость')
plt.ylabel('Вид')
plt.title('Распределение стоимости от вида из окон');
```

На графике видно, что наибольшей популярностью пользуются дома с очень хорошим видом из окна, соответственно с высокой стоимостью
