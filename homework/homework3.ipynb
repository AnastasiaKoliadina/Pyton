1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

````python
df = pd.read_csv('kc_house_data (3).csv',encoding='cp1251')
````

````python
df.head()
````

````python
df.info()
````

````python
# 1.1 Создать новый признак price_per_sq_lot, который будет содержать стоимость за один кв. метр общей площади
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
df.head()
````

````python
# 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0
yr_diff = []

for i, row in df.iterrows():
    if row['yr_renovated']:
        yr_diff.append(row['yr_renovated']-row['yr_built'])
    else:
        yr_diff.append(0)
        
len(yr_diff)
````

````python
df['delta_renovated'] = yr_diff

df[['yr_built', 'yr_renovated', 'delta_renovated']].head(10)
````

````python
# 1.3 Создайте признаки года продажи, месяца продажи
df['year_sale'] = 0
````

````python
df['date'].value_counts()
````

````python
def date(stroka):
    return int(stroka.split()[0][:-11])
````

````python
date('20150427T000000')
````

````python
df['year_sale'] = df['date'].apply(date)
````

````python
df.head()
````

````python
df['month_sale'] = 0
````

````python
def month(stroka2):
    return int(stroka2.split()[0][4:-9])
````

````python
month('20150427T000000')
````

````python
df['month_sale'] = df['date'].apply(month)
````

````python
# 1.4 Удалите признаки date, zipcode, lat, long
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace= True)
````
